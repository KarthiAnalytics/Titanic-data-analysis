```python
import pandas as pd
import numpy as np
```

```python
# Importing csv file into Dataframe
url='https://drive.google.com/uc?export=download&id=1RTVMjhnw6EPltoDa1r72X4HyutHwpT4E'
DF=pd.read_csv(url)
DF
```

```python
#Displaying the top part of the Dataframe
DF.head()
```

```python
#Displaying the bottom part of the Dataframe
DF.tail()
```

```python
#Displaying a summary of the DataFrame
DF.info()
```

```python
#Descriptive statistics for the DataFrame
DF.describe()
```

```python
#Replacing null value of Age column with median
median_age = DF['Age'].median()
DF = DF.fillna(median_age)
DF.info()
DF.head(10)
```

```python
# Capitalizing values in the 'Sex' column
DF['Sex'] = DF['Sex'].str.capitalize()
DF
```

```python
#Converting Age column values to Integer
DF['Age']=DF['Age'].astype('int32')
```

```python
# Displaying all columns with values
pd.set_option('display.max_columns', None)
print(DF.head())
```

```python
#Checking Cabin column null value count
DF['Cabin'].isnull().sum()
```

```python
# Dropping cabin column, because most of the value is null
DF.drop('Cabin', axis=1, inplace=True)
DF
```

```python
#Dropping Duplicate Values in the Dataframe
DF.drop_duplicates(inplace=True)
DF
```

```python
# New Column creation to check the family size
DF['FamilySize'] = DF['SibSp'] + DF['Parch'] + 1
DF.head(2)
```

```python
# Creating a new boolean column 'IsAlone' to check if the person is traveling alone
DF['IsAlone'] = False
DF.loc[DF['FamilySize'] == 1, 'IsAlone'] = True
DF.head()
```

```python
#Splitting FirstName, MiddleName, LastName and Title from FullName and reordering
DF.loc[:,'Title'] = DF['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())
DF.loc[:,'LastName'] = DF['Name'].apply(lambda name: name.split(',')[0].strip())
DF.loc[:,'FirstName'] = DF['Name'].apply(lambda name: name.split(',')[1].split('.',1)[1].strip())
DF.loc[:,'MiddleName'] = DF['FirstName'].apply(lambda x: ' '.join(x.split(' ')[1:]) if len(x.split(' ')) > 1 else '')
DF.loc[:,'FirstName'] = DF['FirstName'].apply(lambda x: x.split(' ')[0])
columns = list(DF.columns)
new_order = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Title', 'FirstName', 'MiddleName','LastName', 'Sex', 'Age', 'SibSp', 'Parch', 'FamilySize', 'IsAlone', 'Ticket', 'Fare', 'Embarked']
DF = DF[new_order]
DF.drop('Name', axis=1, inplace=True)
DF['Ticket'] = DF['Ticket'].str.replace(r'[^0-9]*', '', regex=True)
DF.head()
```

```python
# New column creation Ischild by checking Age column
DF.loc[:, 'IsChild'] = DF['Age'] < 16
DF
```

```python
# New column 'Fareperperson' creation by dividing 'Fare' and 'FamilySize' column
DF.loc[:,'FarePerPerson'] = (DF['Fare'] / DF['FamilySize'].astype('float32')).round(2)
DF.loc[:, 'Fare'] = DF['Fare'].round(2)
DF.head()
```

```python
# Checking survival rate by sex by grouping Sex and survived column with mean value
survival_rate_by_sex = DF.groupby('Sex')['Survived'].mean()
print(survival_rate_by_sex)
```

```python
## Checking survival rate by pclass sex by grouping Pclass,Sex and survived column with mean value
survival_rate_by_pclass_sex = DF.groupby(['Pclass', 'Sex'])['Survived'].mean()
print(survival_rate_by_pclass_sex)
```

```python
#Checking family size counts by using Familysize column value counts
family_size_counts = DF['FamilySize'].value_counts().sort_index()
print(family_size_counts)
```

```python
#Checking class_stats by grouping Pclass, Age and Fare column with mean value
class_stats = DF.groupby('Pclass')[['Age', 'Fare']].mean()
print(class_stats)
```

```python
# Pivot table creation for survival analysis
survival_pivot = DF.pivot_table(
    values='Survived',
    index='Sex',
    columns='Pclass',
    aggfunc='mean'
)
print(survival_pivot)
```
